```python
import numpy as np
rng = np.random.default_rng()
r = rng.normal(size=(100,3))
print(r.shape)
H, edges = np.histogramdd(r, bins = (5, 8, 4))
print(H.shape, edges[0].size, edges[1].size, edges[2].size)
```

```python
import numpy as np

data = np.random.randn(1000, 3)
print(data.shape)
hist, bin_edges = np.histogramdd(data, bins=[4, 4, 4])
print(np.sum(hist))
print("ヒストグラム: ", hist)
print("ビンのエッジ: ", bin_edges)
b=np.array(bin_edges)
print(b.shape)
c=(b[0][1:] + b[0][:-1])/2
print(c)
```

```python
import numpy as np
b=np.array([1, 2, 3])
#b[1]とb[0]の平均、b[2]とb[1]の平均...と計算する
c=[(b[1:] + b[:-1])/2]
#b[-1]は末尾の要素を取るがb[:-1]は末尾から2番目までをとる
print(b[1:], b[:-1])
print(b[-1])
print(c)

#ndarrayにe[None]とすることで外側の[]を一つ増やせる
e=np.array([[1, 2, 3], [4, 5, 6]])
print(e)
d=e[None].astype(np.uint8)
print(d)
print(type(d))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

#描画エリアの作成
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

#numpyを使ってXYZの値を設定
x = np.random.rand(200,1) 
y = np.random.rand(200,1)
z = np.random.rand(200,1)

#散布図の作成
ax.scatter(x,y,z,s=40,c="red")

#描画
plt.show()
```

```python
import numpy as np

# 配列を定義
array1 = np.arange(10).reshape(5, 2)
array2 = np.zeros([5, 3])
array3 = np.arange(5).reshape(5, 1)
# 配列を結合
combined_array = np.concatenate((array1, array2, array3), axis=1)

print(combined_array)
```

```python
import numpy as np
colors = np.random.rand(5, 3)  # 各点のRGB値
print(colors)
# RGB値を16進数に変換
#S.formatで文字列Sの｛｝内に変数入れられる
color_hex = ['rgb({},{},{})'.format(r, g, b) for r, g, b in colors]
print(color_hex)
```
